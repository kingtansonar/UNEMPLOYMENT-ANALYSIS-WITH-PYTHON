```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import pearsonr

# Step 1: Data Collection and Preparation
data = pd.read_csv('/kaggle/input/datasetemp/Unemployment_Rate_upto_11_2020.csv')

# Step 2: Data Cleaning
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)

# Step 3: Data Exploration
print(data.describe())

# Step 4: Time Series Analysis (Optional)
# Remove this step since you're not using the date column

# Step 5: Statistical Analysis
correlation, p_value = pearsonr(data[' Estimated Unemployment Rate (%)'], data[' Estimated Employed'])
print(f'Correlation: {correlation}\nP-value: {p_value}')

# Step 6: Regression Analysis
X = data[[' Estimated Employed', ' Estimated Labour Participation Rate (%)']]
X = sm.add_constant(X)
y = data[' Estimated Unemployment Rate (%)']

model = sm.OLS(y, X).fit()
print(model.summary())

# Step 7: Data Visualization
plt.scatter(data[' Estimated Employed'], data[' Estimated Unemployment Rate (%)'])
plt.plot(data[' Estimated Employed'], model.predict(X), color='red')
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Employed vs Unemployment Rate')
plt.grid(True)
plt.show()

# Step 8: Interpretation and Reporting
# Provide interpretations of the analysis results and prepare a report or presentation.
```
